```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Carregando o CSV

```python
df = pd.read_csv("ddf--datapoints--literacy_rate_youth--by--geo--gender--time.csv")
```

# Seleção de países 

```python
paises = ['bra', 'chn', 'ind', 'rus']
df = df[df['geo'].isin(paises)]
```

# Criando gráfico
## Informações do gráfico:
<ol>
    <li>O eixo x é o ano (df['time']).</li>
    <li>O eixo y é a taxa de alfabetização de adultos (df['literacy_rate_adult']).</li>
    <li>O tamanho das bolinhas é definido usando df['geo'].map(tamanho_bolinha), que usa o dicionário tamanho_bolinha para determinar o tamanho da bolinha para cada país.</li>
    <li>A cor das bolinhas é definida usando df['gender'].map(cores), que usa o dicionário cores para determinar a cor da bolinha para cada gênero.</li>
</ol>

```python
# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))

# Definindo cores para cada gênero
cores = {0: 'pink', 1: 'blue'}

# Criando o gráfico de dispersão para cada país
for pais in paises:
    df_pais = df[df['geo'] == pais]
    plt.scatter(df_pais['time'], df_pais['literacy_rate_youth'], 
                s=df_pais['literacy_rate_youth'] * 10, 
                c=df_pais['gender'].map(cores), 
                label=pais, alpha=0.7)

# Adicionando rótulos e legendas
plt.xlabel('Ano')
plt.ylabel('Taxa de Alfabetização da Juventude')
plt.title('Taxa de Alfabetização da Juventude por País e Gênero ao Longo do Tempo')
plt.legend()

# Criando a barra de cores
colorbar = plt.colorbar()
colorbar.set_label('Gênero (0: Feminino, 1: Masculino)')
colorbar.set_ticks([0, 1])
colorbar.set_ticklabels(['Feminino', 'Masculino'])

# Exibindo o gráfico
plt.show()
```
